#### 2. Read all product sales data and display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product ).
**Note:**It must be implemented in three methods  
The graph should look like this.


```python
plt.figure(figsize=(6,5))
plt.plot(df['month_number'], df['facecream'],   label = 'Face cream Sales Data', marker='o', linewidth=3)  # Plot the chart

plt.plot(df['month_number'], df['facewash'],   label = 'Face Wash Sales Data', marker='o', linewidth=3)  # Plot the chart
plt.plot(df['month_number'], df['toothpaste'],   label = 'ToothPaste Sales Data', marker='o', linewidth=3)  # Plot the chart
plt.plot(df['month_number'], df['bathingsoap'],   label = 'bathingsoap Sales Data', marker='o', linewidth=3)  # Plot the chart
plt.plot(df['month_number'], df['shampoo'],   label = 'shampoo Sales Data', marker='o', linewidth=3)  # Plot the chart
plt.plot(df['month_number'], df['moisturizer'],   label = 'moisturizer Sales Data', marker='o', linewidth=3)  # Plot the chart

plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.legend(loc='upper left')
plt.xticks(np.arange(1,13))
plt.yticks([1000, 2000, 4000, 6000, 8000, 10000, 12000, 15000, 18000])
plt.title('Sales data')
plt.show()
```

#### 5. Read the total profit of each month and show it using the histogram to see the most common profit ranges.  
**Note:**It must be implemented in three methods.  
The histogram should look like this.


```python
profit_range = [150000, 175000, 200000, 225000, 250000, 300000, 350000]
plt.hist(df['total_profit'], profit_range, label = 'Profit data')
plt.xlabel('profit range in dollar')
plt.ylabel('Actual Profit in dollar')
plt.legend(loc='upper left')
plt.xticks(profit_range)
plt.title('Profit data')
plt.show()
```

```python
####################################### histogram Plot using Matplotlib visualization #########################################
```

```python
profit_range = [150000, 175000, 200000, 225000, 250000, 300000, 350000]
sns.histplot(data=df, x='total_profit',bins=profit_range,edgecolor="none")
plt.xlabel('profit range in dollar')
plt.ylabel('Actual Profit in dollar')
plt.xticks(profit_range)
plt.title('Profit data')
```

```python
####################################### histogram  Plot using Seaborn visualization #########################################
```

```python
df['total_profit'].hist(bins=profit_range,color='tab:blue')
plt.grid(False)
plt.xlabel('profit range in dollar')
plt.ylabel('Actual Profit in dollar')
plt.xticks(profit_range)
plt.title('Profit data')
```

#### 6. Calculate total sale data for last year for each product and show it using a Pie chart.
- **Note:**
    - In Pie chart display Number of units sold per year for each product in percentage.
    - It must be implemented in two methods.

The Pie chart should look like this.


```python

labels = ['FaceCream', 'FaseWash', 'ToothPaste', 'Bathing soap', 'Shampoo', 'Moisturizer']
salesData   = [df ['facecream'].sum(), df ['facewash'].sum(), df ['toothpaste'].sum(), 
         df ['bathingsoap'].sum(), df ['shampoo'].sum(), df ['moisturizer'].sum()]
plt.axis("equal")
plt.pie(salesData, labels=labels, autopct='%1.1f%%')
plt.legend(loc='lower right')
plt.title('Sales data')
plt.show()
```

```python
#################################### pie Plot using Matplotlib visualization #####################################
```

```python

df[['facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer']].sum(axis=0).plot.pie( label='',figsize=(11, 6), autopct='%1.1f%%').legend(loc='lower right')
```
